```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('./DATASET/VHM.csv')
# Descriptive statistics
print(df.describe(include='all'))
# Convert close value to numeric
# df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
df['Close']
```

```python
def remove_commas_and_convert(x):
    return float(x.replace(',', ''))
```

```python
df['Close'] = df['Close'].apply(remove_commas_and_convert)
```

```python
print('Mean :',df['Close'].mean())
print('Median :',df['Close'].median())
print('Std:',df['Close'].std())
print('Min:',df['Close'].min())
print('Max:',df['Close'].max())
print('25%',df['Close'].quantile(0.25))
print('50%',df['Close'].quantile(0.5))
print('75%',df['Close'].quantile(0.75))
print('Skewness',df['Close'].skew())
print('Kurtosis',df['Close'].kurtosis())
```

```python
import matplotlib.pyplot as plt
# Time series plot
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df.set_index('Date', inplace=True)
df['Close'].plot()
plt.show()
```
