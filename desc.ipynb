```python
import pandas as pd
import numpy as np
df = pd.read_csv('./DATASET/VHM.csv')
# Descriptive statistics
print(df.head())
# print(df.describe(include='all'))
# Convert close value to numeric
df['Close'] = df['Close'].str.replace(',', '')
df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
# Mean of close values
print(df['Close'].mean())
# Number of observations
print(df['Close'].size)
# Median of close values
print(df['Close'].median())
# Standard deviation of close values
print(df['Close'].std())
# Variance of close values
print(df['Close'].var())
# Minimum of close values
print(df['Close'].min())
# Maximum of close values   
print(df['Close'].max())
# Skewness of close values
print(df['Close'].skew())
# Kurtosis of close values
print(df['Close'].kurt())
# Quantile of close values
print(df['Close'].quantile([0.25, 0.5, 0.75]))
# Range of close values
print(df['Close'].max() - df['Close'].min())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the plots
sns.set_style("whitegrid")

# Time series plot
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df.set_index('Date', inplace=True)

plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Date", y="Close")
plt.title('Time Series Plot of VHM Close Value')
plt.xlabel('Date')
plt.ylabel('Close Value')
plt.show()

# Histogram of close value
plt.figure(figsize=(10,6))
sns.histplot(data=df, x="Close", bins=30, kde=True)
plt.title('Histogram of VHM Close Value')
plt.xlabel('Close Value')
plt.ylabel('Frequency')
plt.show()

# Boxplot of close value
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Close")
plt.title('Boxplot of VHM Close Value')
plt.xlabel('Close Value')
plt.show()
```
